
# Task 2: Read 2 images
Calculate the Euclidean distance between 2 images: use the spicy library. 
Calculate the cosine similarity between the 2 images.

```python
path_14 = os.path.join(dir_images,'image_0814.jpg')
path_16 = os.path.join(dir_images,'image_0816.jpg')
path_48 = os.path.join(dir_images,'image_1248.jpg')
```

```python
def TinhHist(pathfilename):
    img = cv.imread(pathfilename,0)
    hist = cv.calcHist([img],[0],None,[256],[0,256])
    size = img.shape[0]*img.shape[1]
    hist = hist / size
    return hist.flatten()
```

```python
hist_14 = TinhHist(path_14)
hist_16 = TinhHist(path_16)
hist_48 = TinhHist(path_48)
```

```python
from scipy.spatial import distance
d14_16 = distance.euclidean(hist_14, hist_16)
d14_48 = distance.euclidean(hist_14, hist_48)
```

```python
print("The Euclidean distance between images 14 and 16 is: ", d14_16)
print("The Euclidean distance between images 14 and 48 is: ", d14_48)
```

```python
cosine_14_16 = np.dot(hist_16,hist_14)/(norm(hist_16)*norm(hist_14))
cosine_14_48 = np.dot(hist_48,hist_14)/(norm(hist_48)*norm(hist_14))
```

```python
print("The cosine similarity between images 14 and 16:", cosine_14_16)
print("The cosine similarity between images 14 and 48:", cosine_14_48)
```
