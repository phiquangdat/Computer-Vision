```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from random import randint
```

```python
def find_max(R, i, j):
 max_point = 0
 
 for k in [-1, 0, 1]:
     for l in [-1, 0, 1]:
         if (max_point < R[i+k][j+l]):
             max_point = R[i+k][j+l]
 return max_point
```

```python
def matrix_multiply(I_x, I_y):
 (h, w) = I_x.shape
 result = np.empty((h, w))
 
 for i in range(0, h):
     for j in range(0, w):
         result[i][j] = I_x[i][j] * I_y[i][j]

 return result
```

```python
def nonmax_suppression(R, i, j, max_point):
 for k in [-1, 0, 1]:
     for l in [-1, 0, 1]:
         if (R[i+k][j+l] == max_point):
             continue
         else:
             R[i+k][j+l] = 0
```

```python
def calculate_trace(A, B):
 return A + B
```

```python
def calculate_det(A, B, C):
 det = matrix_multiply(A, B) - matrix_multiply(C, C)
 
 return det
```

```python
img1 = cv2.imread(cow1)
img2 = cv2.imread(cow2)
img3 = cv2.imread(graf_img1)
img4 = cv2.imread(graf_img5)
img5 = cv2.imread(match1)
img6 = cv2.imread(match2)
```

```python
gaussian_blur1 = cv2.GaussianBlur(cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY),(5,5),2)
gaussian_blur2 = cv2.GaussianBlur(cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY),(5,5),2)
gaussian_blur3 = cv2.GaussianBlur(cv2.cvtColor(img3, cv2.COLOR_BGR2GRAY),(5,5),2)
gaussian_blur4 = cv2.GaussianBlur(cv2.cvtColor(img4, cv2.COLOR_BGR2GRAY),(5,5),2)
gaussian_blur5 = cv2.GaussianBlur(cv2.cvtColor(img5, cv2.COLOR_BGR2GRAY),(5,5),2)
gaussian_blur6 = cv2.GaussianBlur(cv2.cvtColor(img6, cv2.COLOR_BGR2GRAY),(5,5),2)
```

```python
I_X1 = cv2.Sobel(gaussian_blur1, cv2.CV_64F, 1, 0, ksize = 5)
I_Y1 = cv2.Sobel(gaussian_blur1, cv2.CV_64F, 0, 1, ksize = 5)
I_X2 = cv2.Sobel(gaussian_blur2, cv2.CV_64F, 1, 0, ksize = 5)
I_Y2 = cv2.Sobel(gaussian_blur2, cv2.CV_64F, 0, 1, ksize = 5)
I_X3 = cv2.Sobel(gaussian_blur3, cv2.CV_64F, 1, 0, ksize = 5)
I_Y3 = cv2.Sobel(gaussian_blur3, cv2.CV_64F, 0, 1, ksize = 5)
I_X4 = cv2.Sobel(gaussian_blur4, cv2.CV_64F, 1, 0, ksize = 5)
I_Y4 = cv2.Sobel(gaussian_blur4, cv2.CV_64F, 0, 1, ksize = 5)
I_X5 = cv2.Sobel(gaussian_blur5, cv2.CV_64F, 1, 0, ksize = 5)
I_Y5 = cv2.Sobel(gaussian_blur5, cv2.CV_64F, 0, 1, ksize = 5)
I_X6 = cv2.Sobel(gaussian_blur6, cv2.CV_64F, 1, 0, ksize = 5)
I_Y6 = cv2.Sobel(gaussian_blur6, cv2.CV_64F, 0, 1, ksize = 5)
```

```python
A1 = matrix_multiply(I_X1, I_X1)
B1 = matrix_multiply(I_X1, I_Y1)
C1 = matrix_multiply(I_Y1, I_Y1)
A2 = matrix_multiply(I_X2, I_X2)
B2 = matrix_multiply(I_X2, I_Y2)
C2 = matrix_multiply(I_Y2, I_Y2)
A3 = matrix_multiply(I_X3, I_X3)
B3 = matrix_multiply(I_X3, I_Y3)
C3 = matrix_multiply(I_Y3, I_Y3)
A4 = matrix_multiply(I_X4, I_X4)
B4 = matrix_multiply(I_X4, I_Y4)
C4 = matrix_multiply(I_Y4, I_Y4)
A5 = matrix_multiply(I_X5, I_X5)
B5 = matrix_multiply(I_X5, I_Y5)
C5 = matrix_multiply(I_Y5, I_Y5)
A6 = matrix_multiply(I_X6, I_X6)
B6 = matrix_multiply(I_X6, I_Y6)
C6 = matrix_multiply(I_Y6, I_Y6)
```

```python
_A1 = cv2.GaussianBlur(A1, (5,5), 2)
_B1 = cv2.GaussianBlur(B1, (5,5), 2)
_C1 = cv2.GaussianBlur(C1, (5,5), 2)
_A2 = cv2.GaussianBlur(A2, (5,5), 2)
_B2 = cv2.GaussianBlur(B2, (5,5), 2)
_C2 = cv2.GaussianBlur(C2, (5,5), 2)
_A3 = cv2.GaussianBlur(A3, (5,5), 2)
_B3 = cv2.GaussianBlur(B3, (5,5), 2)
_C3 = cv2.GaussianBlur(C3, (5,5), 2)
_A4 = cv2.GaussianBlur(A4, (5,5), 2)
_B4 = cv2.GaussianBlur(B4, (5,5), 2)
_C4 = cv2.GaussianBlur(C4, (5,5), 2)
_A5 = cv2.GaussianBlur(A5, (5,5), 2)
_B5 = cv2.GaussianBlur(B5, (5,5), 2)
_C5 = cv2.GaussianBlur(C5, (5,5), 2)
_A6 = cv2.GaussianBlur(A6, (5,5), 2)
_B6 = cv2.GaussianBlur(B6, (5,5), 2)
_C6 = cv2.GaussianBlur(C6, (5,5), 2)
```

```python
K = 0.04
```

```python
R1 = calculate_det(_A1, _B1, _C1) - K*calculate_trace(_A1, _B1)
R2 = calculate_det(_A2, _B2, _C2) - K*calculate_trace(_A2, _B2)
R3 = calculate_det(_A3, _B3, _C3) - K*calculate_trace(_A3, _B3)
R4 = calculate_det(_A4, _B4, _C4) - K*calculate_trace(_A4, _B4)
R5 = calculate_det(_A5, _B5, _C5) - K*calculate_trace(_A5, _B5)
R6 = calculate_det(_A6, _B6, _C6) - K*calculate_trace(_A6, _B6)
```

```python
_, R1 = cv2.threshold(R1, R1.max() // 500, R1.max(), cv2.THRESH_BINARY)
_, R2 = cv2.threshold(R2, R2.max() // 500, R2.max(), cv2.THRESH_BINARY)
_, R3 = cv2.threshold(R3, R3.max() // 500, R3.max(), cv2.THRESH_BINARY)
_, R4 = cv2.threshold(R4, R4.max() // 500, R4.max(), cv2.THRESH_BINARY)
_, R5 = cv2.threshold(R5, R5.max() // 500, R5.max(), cv2.THRESH_BINARY)
_, R6 = cv2.threshold(R6, R6.max() // 500, R6.max(), cv2.THRESH_BINARY)
```

```python
(h, w) = R1.shape

for i in range(1, h-1, 2):
  for j in range(1, w-1, 2):
    max_pixel1 = find_max(R1,i,j)
    nonmax_suppression(R1, i, j, max_pixel1)
(h, w) = R2.shape

for i in range(1, h-1, 2):
  for j in range(1, w-1, 2):
    max_pixel2 = find_max(R2,i,j)
    nonmax_suppression(R2, i, j, max_pixel2)
(h, w) = R3.shape

for i in range(1, h-1, 2):
  for j in range(1, w-1, 2):
    max_pixel3 = find_max(R3,i,j)
    nonmax_suppression(R3, i, j, max_pixel3)
(h, w) = R4.shape

for i in range(1, h-1, 2):
  for j in range(1, w-1, 2):
    max_pixel4 = find_max(R4,i,j)
    nonmax_suppression(R4, i, j, max_pixel4)
(h, w) = R5.shape

for i in range(1, h-1, 2):
  for j in range(1, w-1, 2):
    max_pixel5 = find_max(R5,i,j)
    nonmax_suppression(R5, i, j, max_pixel5)
(h, w) = R6.shape

for i in range(1, h-1, 2):
  for j in range(1, w-1, 2):
    max_pixel6 = find_max(R6,i,j)
    nonmax_suppression(R6, i, j, max_pixel6)
```

```python
R1_dst = cv2.dilate(R1, None)
R2_dst = cv2.dilate(R2, None)
R3_dst = cv2.dilate(R3, None)
R4_dst = cv2.dilate(R4, None)
R5_dst = cv2.dilate(R5, None)
R6_dst = cv2.dilate(R6, None)
```
